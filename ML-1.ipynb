```python
data=data.drop(['S.No.'], axis = 1)
data.info()
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
data['Brand']=data.Name.str.split().str.get(0)
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2) + data.Name.str.split().str.get(3)
data[['Name','Brand','Model']]
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land','Hindustan']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
```

```python
search_for= ['Hindustan']
data[data.Brand.str.contains('|'.join(search_for))].head(5)
```

```python
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover","Hindustan":"Hindustan Motors"}, inplace=True)
```

```python
plt.figure(figsize=(10,6))
x=['Maruti','Hyundai','Honda', 'Toyota','Mercedes-Benz']
y=[1444,1340,743,507,380]
plt.bar(x,y,color="yellow",edgecolor="black")
plt.xlabel("Brand")
plt.ylabel("No of Cars available")
plt.title("No. of Cars by Brand")
plt.savefig("D:\AIML graphs\graph4.png")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,8))
colors=['yellow','red','pink','orange','brown']
x=np.array([351,331,312,285,228])
labels=['Ford','Mahindra','BMW','Audi','Tata']
plt.pie(x,labels=labels,autopct='%1.1f%%',startangle=140,colors=colors)
plt.title('Car Brand Distribution (Pie Chart)')
plt.axis('equal')  # Equal aspect ratio to make the pie circular
plt.tight_layout()
plt.savefig("D:\AIML graphs\graph3.png")
plt.show()
```
