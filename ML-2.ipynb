```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
print(x)
print(y)
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")
```

```python
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```

```python
y_pred = model.intercept_ + model.coef_ * x
print(f"predicted response:\n{y_pred}")
```

```python
data= pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.describe()
```

```python
plt.scatter(data['YearsExperience'], data['Salary'], color='purple')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X = data.iloc[:, :1] 
y = data.iloc[:, 1:] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
linearModel= LinearRegression()
linearModel.fit(X_train, y_train)
```

```python
y_pred_test = linearModel.predict(X_test)    
y_pred_train = linearModel.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='pink')
plt.show()
```

```python
print(f'Coefficient: {linearModel.coef_}')
print(f'Intercept: {linearModel.intercept_}')
```
